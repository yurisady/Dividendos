```python
import pandas as pd
import yfinance as yf
```

```python
#Importando os dados históricos gerais

file_names = ['ativos_2007.csv', 'ativos_2008.csv', 'ativos_2009.csv', 'ativos_2010.csv', 'ativos_2011.csv', 'ativos_2012.csv', 'ativos_2013.csv', 'ativos_2014.csv', 'ativos_2015.csv', 'ativos_2016.csv', 'ativos_2017.csv', 'ativos_2018.csv', 'ativos_2019.csv', 'ativos_2020.csv', 'ativos_2021.csv', 'ativos_2022.csv']

for file in file_names:
    list_name = file.split('.')[0]
    df = pd.read_csv(file)
    lista = df.values.tolist()
    flat_list = []
    for sublist in lista:
        for item in sublist:
            flat_list.append(item)
    locals()[list_name] = flat_list
```

```python
#Tratando os dados de 2007

lista_2007 = []
for a in ativos_2007:
  print('trabalhando no ativo = ',a)
  try:
    dados = yf.download(tickers=a,actions=True,start='2007-01-02',end='2007-12-31')
    dp = (dados['Close'][-1]-dados['Open'][0])/dados['Open'][0]*100
    d = dados['Dividends']
    freq = d[d != 0].count()
    td = d.sum()
    pm = (dados['Close'].mean())
    dyield = (td/pm)
    ticker = yf.Ticker(a)
    info = ticker.info
    setor = info['sector']
    lista_2007.append([a, td, dp, freq, pm, dyield, setor])
  except:
    # Se ocorrer um erro, imprima uma mensagem e pule para o próximo ativo
    print(f"Erro ao processar o ativo {a}, pulando para o próximo ativo")
  continue

df2007 = pd.DataFrame(lista_2007, columns=['Ativo','SomaDividendos','DeltaP','Freq','PrecoMedio','Dividend Yield', 'Setor'])
```

```python
df2007
```

```python
# Seleciona os ativos cujo Dividend Yield é maior que zero

df2007_filtrado = df2007.loc[(df2007['Dividend Yield'] > 0)]
df2007_filtrado.sort_values(by='Dividend Yield', ascending=False)
```

```python
df2007_filtrado[['Ativo']].to_csv('ativos_div_2007.csv', index=False)
```
